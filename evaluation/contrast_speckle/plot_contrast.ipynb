```python
import matplotlib.pyplot as plt
```

```python
x1 = [1,2,3,4,5]

y_00 = [8.80, 10.5, 10.7, 11.6, 12.2]
y_10 = [10.00, 11.7, 12.1, 12.5, 13.3]
y_20 = [9.40, 11.2, 11.7, 12.1, 12.6]

y_01 = [9.60, 10.7, 11.5, 12.1, 12.3]
y_11 = [10.5, 12, 13.2, 13.3, 14]
y_21 = [10.5, 12.1, 13.1, 12.9, 14]

y_02 = [8.8, 10.5, 11.3, 11.6, 11.6]
y_12 = [10.2, 11, 12.1, 12.7, 12.3]
y_22 = [9.7, 10.9, 11.9, 12.4, 12.3]


z_00 = [8.80, 10.5, 10.7, 10.9, 9.4]
z_10 = [10.00, 11.7, 12.1, 12.1, 9.7]
z_20 = [9.40, 11.2, 11.7, 11.2, 10.7]

z_01 = [9.60, 10.7, 11.5, 9.8, 9.1]
z_11 = [10.5, 12, 13.2, 11.6, 11.3]
z_21 = [10.5, 12.1, 13.1, 12.8, 12.7]

z_02 = [8.8, 10.5, 11.3, 8.9, 9.4]
z_12 = [10.2, 11, 12.1, 9.7, 9.8]
z_22 = [9.7, 10.9, 11.9, 11.5, 11.5]

a_00 = [8.80, 10.5, 10.7, 11.6, 10.9]
a_10 = [10.00, 11.7, 12.1, 12.3, 11.6]
a_20 = [9.40, 11.2, 11.7, 11.8, 11.6]

a_01 = [9.60, 10.7, 11.5, 11.2, 11.1]
a_11 = [10.5, 12, 13.2, 13.2, 13.5]
a_21 = [10.5, 12.1, 13.1, 13.6, 13.9]

a_02 = [8.8, 10.5, 11.3, 10.5, 11.2]
a_12 = [10.2, 11, 12.1, 11.4, 12.0]
a_22 = [9.7, 10.9, 11.9, 11.6, 12.3]
```

```python
plt.figure(figsize=[10,10])
plt.plot(x1, y_00, label='lt greedy')
plt.plot(x1, y_10, label='lm greedy')
plt.plot(x1, y_20,  label='lb greedy')

plt.plot(x1, z_00, label='lt bad strat')
plt.plot(x1, z_10, label='lm bad strat')
plt.plot(x1, z_20, label='lb bad strat')

plt.plot(x1, a_00, label='lt good strat')
plt.plot(x1, a_10, label='lm good strat')
plt.plot(x1, a_20, label='lb good strat')
plt.legend()
```

```python
plt.figure(figsize=[10,10])
plt.plot(x1, y_01,  label= 'mt greedy')
plt.plot(x1, y_11,label='mm greedy')
plt.plot(x1, y_21, label='mb greedy')

plt.plot(x1, z_01, label='mt bad strat')
plt.plot(x1, z_11, label='mm bad strat')
plt.plot(x1, z_21, label='mb bad strat')

plt.plot(x1, a_01, label='mt good strat')
plt.plot(x1, a_11, label='mm good strat')
plt.plot(x1,a_21, label='mb good strat')
plt.legend() 
```
